```python
import pandas as pd
import numpy as np
import sklearn.metrics
```

```python
df = pd.read_csv('../data/raw/2022_train.csv')
```

```python
df_cleaned = df.copy()
```

```python
df_cleaned.drop('Id', axis=1, inplace=True)
df_cleaned.drop('FGM', axis=1, inplace=True)
df_cleaned.drop('FGA', axis=1, inplace=True)
df_cleaned.drop('3P Made', axis=1, inplace=True)
df_cleaned.drop('3PA', axis=1, inplace=True)
df_cleaned.drop('FTA', axis=1, inplace=True)
df_cleaned.drop('FTM', axis=1, inplace=True)
df_cleaned.drop('GP', axis=1, inplace=True)
```

```python
df_cleaned.describe()
```

```python
target = df_cleaned.pop('TARGET_5Yrs')
```

```python
df_cleaned.head()
```

```python
# Standarize Data
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
df_cleaned = scaler.fit_transform(df_cleaned)
```

```python
#df_cleaned
```

```python
# Save standarized data
from joblib import dump
```

```python
dump(scaler, '../models/scaler.joblib')
```

```python
# Using Skicit-learn to split data into training and validation sets
from sklearn.model_selection import train_test_split
```

```python
# Split the data into training and validation sets
X_train, X_val, y_train, y_val = train_test_split(df_cleaned, target, test_size = 0.2, random_state = 7)
```

```python
#Save sets into Data/Processed Folder
np.save('../data/processed/X_train', X_train)
np.save('../data/processed/X_val', X_val)
np.save('../data/processed/y_train', y_train)
np.save('../data/processed/y_val', y_val)
```

```python
#Get Baseline
y_mean = y_train.mean()
```

```python
y_mean
```

```python
y_base = np.full((len(y_train),1), y_mean)
```

```python
from sklearn.metrics import mean_squared_error as mse
from sklearn.metrics import mean_absolute_error as mae
```

```python
print(mse(y_train, y_base, squared=False))
print(mae(y_train, y_base))
```
